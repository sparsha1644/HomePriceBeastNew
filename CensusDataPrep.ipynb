```python
import pandas as pd
```

```python
data_folder = "C:\\Users\\spars\\Documents\\Master\\JHU\\TML\\HomePriceBeastNew\\"
```

```python
education_data = pd.read_csv(f"{data_folder}Education_data.csv")
education_data = education_data.drop(['Rural-urban_Continuum Code_2003_ED',
       'Urban_Influence_Code_2003_ED', 'Rural-urban_Continuum Code_2013_ED',
       'Urban_Influence_Code_2013_ED', 'State', 'Area name'], axis=1)
education_data.rename(columns = {'FIPS Code' : 'FIPS_Code'}, inplace=True)
education_data.head(3)
```

```python
population_data = pd.read_csv(f"{data_folder}Population_data.csv")
population_data = population_data.drop(['Rural-urban_Continuum Code_2003',
       'Rural-urban_Continuum Code_2013', 'Urban_Influence_Code_2003',
       'Urban_Influence_Code_2013'], axis=1)
population_data.rename(columns = {'FIPStxt' : 'FIPS_Code', 'State':'state_code','Area_Name':'county_name'}, inplace=True)
population_data.head(3)
```

```python
unemployment_data = pd.read_csv(f"{data_folder}Unemployment_data.csv")
unemployment_data = unemployment_data.drop(['Rural_urban_continuum_code_2013',
       'Urban_influence_code_2013', 'Metro_2013', 'State', 'Area_name'], axis=1)
unemployment_data.head(3)
```

```python
def combine_census_datasets(education_data, population_data, unemployment_data):
    
    merged_census_data = pd.merge(
    population_data,
    education_data,
    how="inner",
    on="FIPS_Code")

    merged_census_data = pd.merge(
        merged_census_data,
        unemployment_data,
        how="inner",
        on="FIPS_Code")
    
    return merged_census_data
```

```python
merged_census_data = combine_census_datasets(education_data, population_data, unemployment_data)
```

```python
merged_census_data.head(3)
```

```python
#Filter only county data. 
merged_census_county_data = merged_census_data[merged_census_data['FIPS_Code'] % 10 != 0]
```

```python
merged_census_county_data.to_csv(f"{data_folder}merged_census_county_data.csv", index=False)
```
