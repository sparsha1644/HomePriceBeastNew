```python
def get_lead_lag_features(feature_df,L,colname,time_feature,lag=True):
    col_mid = "lead"
    mult = -1
    
    if lag:
        col_mid = "lag"
        mult = 1
        
    other_cols = [x for x in feature_df.columns if x!=colname and x!=time_feature]
    cols = other_cols+ [time_feature]
    feature_df = feature_df.sort_values(cols).reset_index(drop=True)
    print(feature_df.columns)
    
    new_feature_df = pd.DataFrame()
    for lag in range(L):
        l = lag +1
        col = f"{colname}_{col_mid}_{l}"
        new_feature_df[col] = feature_df.groupby(other_cols)[colname].shift(l*mult)
    new_feature_df.fillna(0,inplace=True)
    
    result = pd.concat([feature_df, new_feature_df], axis=1)
    return result
```

```python
def get_lag_features(feature_df, J, colname, time_feature):
    
    return get_lead_lag_features(feature_df, J, colname, time_feature)

def get_lead_features(feature_df, H, colname, time_feature):
    
    return get_lead_lag_features(feature_df, H, colname, time_feature,lag=False)
```

```python
#y - variable of interest. 
#J - Number of historical points to consider. 
#H - Number of future points to predict. 
#X - Columns names of static features of interest. 
def convert_to_timeseries_df(y, J, H, X, data_frame, x_related, time_lag_features, get_lead_for_related=False):
    static_feature_df = data_frame[X].drop_duplicates()
    group_cols = ["state_code", "county_name"]
    dt_col = "period_begin"
    drop_cols = group_cols + [dt_col]
    
    y_feature_df = data_frame[group_cols+[dt_col,y]].drop_duplicates()
    print(y_feature_df.columns)
    
    x_related_lead_feature_df = pd.DataFrame()
    x_related_feature_df = data_frame[group_cols+[dt_col,x_related]].drop_duplicates()
    
    time_features_df = pd.DataFrame()
    time_features_df = data_frame[group_cols+[dt_col]+time_lag_features].drop_duplicates()

    if get_lead_for_related:
        x_related_lead_feature_df = get_lead_features(x_related_feature_df, H, x_related, dt_col)
        x_related_lead_feature_df = x_related_lead_feature_df[[x for x in x_related_lead_feature_df.columns if "lead" in x]]
    
    all_tf_columns = []
    for time_feature in time_lag_features:
        lag_df = get_lag_features(time_features_df[drop_cols + [time_feature]], J, time_feature, dt_col)
        lag_df.drop(columns=drop_cols, inplace=True)
        all_tf_columns.append(lag_df)
            
    time_lag_features_df = pd.concat(all_tf_columns,axis = 1)  
    x_related_feature_df = get_lag_features(x_related_feature_df, J, x_related, dt_col) 
    print(x_related_feature_df.columns)
    
    #Get lag and lead terms for y_feature
    y_feature_df_lag = get_lag_features(y_feature_df, J, y, dt_col)    
    y_feature_df_lead = get_lead_features(y_feature_df, H, y, dt_col)
    y_feature_df_lead = y_feature_df_lead[[x for x in y_feature_df_lead.columns if x not in y_feature_df_lag.columns]]
    
    y_feature_df = pd.concat([y_feature_df_lag, 
                              y_feature_df_lead], axis=1)
    
    return static_feature_df, x_related_feature_df, y_feature_df, time_lag_features_df, x_related_lead_feature_df
```

```python
time_lag_features = ["week_num", "month", "week_offset"]
static_feature_df, x_related_feature_df, y_feature_df, time_features_df, _ = convert_to_timeseries_df('inventory', J, H, X, home_data, 'median_sale_price', time_lag_features)
```
